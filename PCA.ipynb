```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mglearn

fig, axes = plt.subplots(15, 2, figsize=(10,20)) # 총 30개. 엑시스 각각에 우린 지금 히스토그램할거임. 말리그넌트는 악성 비나인은 양성이라할 수 있음. 말리그넌트 꼭찾아야하는데이터
malignant = cancer.data[cancer.target ==0]
benign = cancer.data[cancer.target == 1]
ax = axes.ravel() # 원래 이중포문 써야하는ㄱ=데 ravel 써서 하나로 쭉 펴서 포문 하나만 써도 되는것임(?)

for i in range(30):
  _, bins = np.histogram(cancer.data[:, i], bins=50) # 빈스가 50인거임


  ax[i].hist(malignant[:, i], bins=bins, color=mglearn.cm3(0), alpha=.5)
  ax[i].hist(benign[:,i], bins=bins, color=mglearn.cm3(2), alpha=.5)
  ax[i].set_title(cancer.feature_names[i])
  ax[i].set_yticks(())
ax[0].set_xlabel("특성 크기")
ax[0].set_ylabel("빈도")
ax[0].legend(['mailgnant', 'benign' ], loc='best')
fig.tight_layout()
```

###zero scaling
- 전처리 : 각 특성의 분산이 1이 되도록 스케일링한다.

```python
from sklearn.preprocessing import StandardScaler
cancer = load_breast_cancer()

scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)
X_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_scaled)
```

```python
X_pca = pca.transform(X_scaled)
print(f'원본 데이터 형태 : {X_scaled.shape}')
print(f'축소된 데이터 형태 : {X_pca.shape}')
```

```python
plt.figure(figsize=(8,8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], cancer.target)
plt.legend(['malignant', 'benign'], loc='best')
plt.gca().set_aspect('equal')
plt.xlabel('첫 번째 주성분')
plt.ylabel('두 번째 주성분')
```

```python
print(f'PCA 주성분 형태 : {pca.components_.shape}, {pca.components_}')
```

```python
plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0, 1], ['첫 번째 주성분', '두 번째 주성분'])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)), cancer.feature_names, rotation=60, ha='left')
plt.xlabel('특성')
plt.ylabel('주성분')
```
