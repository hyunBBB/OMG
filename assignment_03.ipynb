```python
import SimpleITK as sitk
from skimage.transform import rotate
import cv2 as cv
from keras.preprocessing import image as kimg
import numpy as np
```

```python
test_data = './Deepnoid-Education/dataset/test.dcm'
#############################
# fill the blank
# original image 와 augmentation 한 image를 4장을 만드세요 
# 5장의 이미지를 하나의 array로 만드세요
# array를 마지막에 있는 display의 argument로 넣어서 현재 출력된 이미지와 유사하도록 만드세요
origin_image = sitk.ReadImage(test_data)
img_arr = sitk.GetArrayFromImage(origin_image)
img_rot = rotate(img_arr[0], 20)
img_shift = kimg.random_shift(np.array(img_arr), 1, 0.5)
img_hflip = np.fliplr(img_arr[0])
img_vflip = np.flipud(img_arr[0])
img_rot = img_rot[np.newaxis,]
img_hflip = img_hflip[np.newaxis,]
img_vflip = img_vflip[np.newaxis,]
arr = np.array(img_arr)
imgList = [img_rot, img_shift, img_hflip, img_vflip]
for img in imgList:
    arr = np.concatenate((arr, img), axis=0)
display(arr)
```

```python
# display 함수는 수정하지 말고 그대로 사용해주세요.
def display(array):
  import matplotlib.pyplot as plt
  for i in range(array.shape[0]):
    plt.imshow(array[i], cmap='gray')
    plt.show()
```
